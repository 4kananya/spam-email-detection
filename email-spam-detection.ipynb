```python
import numpy as np
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
import nltk
nltk.download('stopwords')
```

```python
# printing the stopwords in English
print(stopwords.words('english'))
```

```python
# loading the dataset to a pandas DataFrame
email_dataset = pd.read_csv('/kaggle/input/email-spam-detection-dataset-classification/spam.csv',encoding='latin-1')
email_dataset.head()
```

```python
# counting the number of missing values in the dataset
email_dataset.isnull().sum()
```

```python
#Let's drop the non-relevant unnamed columns 
email_dataset=email_dataset.drop(['Unnamed: 2', 'Unnamed: 3','Unnamed: 4'],axis=1)
email_dataset=email_dataset.rename(columns={"v1":"Category","v2":"Text"})
email_dataset.head()
```

```python
email_dataset.loc[:,'Category']=email_dataset.Category.map({'ham':0, 'spam':1})
email_dataset['Category'] = email_dataset['Category'].astype(int)
email_dataset.head()
```

```python
#separating the data and label
X = email_dataset['Text'].values
Y = email_dataset['Category'].values
print(X)
print(Y)
```

```python
# converting the textual data to numerical data
vectorizer = TfidfVectorizer()
vectorizer.fit(X)

X = vectorizer.transform(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state=20)
```

```python
# Multinomial Naive Bayes model 
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train, Y_train)  # Train the model

prediction = model.predict(X_test)

print("Multinomial NB")
print("Accuracy score: {}". format(accuracy_score(Y_test, prediction)) )
print("Precision score: {}". format(precision_score(Y_test, prediction)) )
print("Recall score: {}". format(recall_score(Y_test, prediction)))
print("F1 score: {}". format(f1_score(Y_test, prediction)))
```

```python
test_data_accuracy = accuracy_score(prediction, Y_test)
print('Accuracy score of the test data : ', test_data_accuracy)
```

```python
X_new = X_test[3]

prediction = model.predict(X_new)
print(prediction)

if (prediction[0]==0):
  print('The mail is Ham')
else:
  print('The mail is Spam')
```
